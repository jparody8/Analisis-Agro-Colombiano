### **2.3. Valores atípicos**
---
Al abordar valores atípicos, es relevante cuestionarse sobre la naturaleza de estos puntos extremos.




```python
def analizar_outliers(df, columna):
    """Analiza si los outliers son errores o valores válidos excepcionales"""

    # Identificar outliers usando IQR
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers = df[(df[columna] < limite_inferior) | (df[columna] > limite_superior)]

    # Análisis de contexto
    print(f"📊 ANÁLISIS DE OUTLIERS PARA - {columna}")
    print(f"👾 Total outliers: {len(outliers)}")
    print(f"🔣 Porcentaje: {len(outliers)/len(df)*100:.2f}%")

    # Verificar patrones lógicos
    if columna == 'Rendimiento (t/ha)':
        # Rendimientos imposibles (negativos o extremadamente altos)
        errores_logicos = outliers[outliers[columna] <= 0]
        print(f"Rendimientos <= 0 (errores): {len(errores_logicos)}")

        # Rendimientos por cultivo para contexto
        rendimiento_por_cultivo = df.groupby('Cultivo')[columna].agg(['mean', 'std', 'max'])
        outliers_con_cultivo = outliers.merge(rendimiento_por_cultivo, on='Cultivo', suffixes=('', '_promedio'))
```

```python
# Función simplificada para el notebook
def visualizar_outliers_simple(df, columna):

    # Calcular límites
    Q1, Q3 = df[columna].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    limite_inf, limite_sup = Q1 - 1.5*IQR, Q3 + 1.5*IQR

    # Clasificar
    normales = df[(df[columna] >= limite_inf) & (df[columna] <= limite_sup)]
    outliers = df[(df[columna] < limite_inf) | (df[columna] > limite_sup)]
    errores = outliers[outliers[columna] <= 0] if columna == 'Rendimiento (t/ha)' else outliers[outliers[columna] < 0]
    validos = outliers[outliers[columna] > 0] if columna == 'Rendimiento (t/ha)' else outliers[outliers[columna] >= 0]

    # Visualizar
    fig, (ax1) = plt.subplots(1, 1, figsize=(14, 5))

    # Boxplot
    ax1.boxplot([normales[columna].dropna()], labels=['Datos'])
    ax1.scatter([1]*len(errores), errores[columna], color='red', s=30, alpha=0.7, label=f'Errores: {len(errores)}')
    ax1.set_title(f'Outliers: {columna}')
    ax1.legend()



    plt.tight_layout()
    plt.show()

    print(f"📊 RESUMEN - {columna}")
    print(f"🔵 Normales: {len(normales):,} ({len(normales)/len(df)*100:.1f}%)")
    print(f"🔴 Errores: {len(errores):,} ({len(errores)/len(df)*100:.1f}%)")
    print(f"🟠 Excepcionales válidos: {len(validos):,} ({len(validos)/len(df)*100:.1f}%)")
```

```python
def analizar_contexto_regional(df, columna):

    Q1, Q3 = df[columna].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    outliers = df[(df[columna] < Q1 - 1.5*IQR) | (df[columna] > Q3 + 1.5*IQR)]

    # Análisis por departamento
    outliers_por_depto = outliers.groupby('Departamento')[columna].agg(['count', 'mean', 'max']).round(2)
    outliers_por_depto.columns = ['Cantidad_Outliers', 'Rendimiento_Promedio', 'Rendimiento_Máximo']

    print("📊 \nCONTEXTO REGIONAL DE VALORES ATÍPICOS: ")
    print(outliers_por_depto.sort_values('Rendimiento_Máximo', ascending=False).head(10))

    # Análisis por cultivo
    outliers_por_cultivo = outliers.groupby('Cultivo')[columna].agg(['count', 'mean', 'max']).round(2)
    outliers_por_cultivo.columns = ['Cantidad_Outliers', 'Rendimiento_Promedio', 'Rendimiento_Máximo']

    print("📊 \nCULTIVOS CON MAYOR VARIABILIDAD: ")
    print(outliers_por_cultivo.sort_values('Rendimiento_Máximo', ascending=False).head(10))
```

### **2.4. Datos Inconsistentes**
---
Frente a inconsistencias en los datos, es importante analizar cómo estas discrepancias podrían afectar la comparabilidad y comprensión de la información.

* ¿Cómo se identificaron las inconsistencias en los datos?-

Entendiendo los datos y el comportamiento logico de las variables






```python
def identificar_inconsistencias_logicas(df):
    """Identifica inconsistencias en los datos del dataframe agrícola"""

    print("🔍 IDENTIFICACIÓN DE INCONSISTENCIAS")
    print("=" * 50)

    inconsistencias = {}

    # 1. Inconsistencias matemáticas
    if all(col in df.columns for col in ['Área cosechada (ha)', 'Producción (t)', 'Rendimiento (t/ha)']):
        df['rendimiento_calculado'] = df['Producción (t)'] / df['Área cosechada (ha)']
        df['diferencia_rendimiento'] = abs(df['Rendimiento (t/ha)'] - df['rendimiento_calculado'])

        # Tolerancia de 0.1 t/ha
        incons_matematicas = df[df['diferencia_rendimiento'] > 0.1]
        inconsistencias['matematicas'] = len(incons_matematicas)

        print(f"📊 Inconsistencias matemáticas: {len(incons_matematicas):,} registros")
        print(f"   Porcentaje: {len(incons_matematicas)/len(df)*100:.2f}%")

    # 2. Valores imposibles
    valores_negativos = df[(df['Área sembrada (ha)'] < 0) |
                          (df['Área cosechada (ha)'] < 0) |
                          (df['Producción (t)'] < 0) |
                          (df['Rendimiento (t/ha)'] <= 0)]
    inconsistencias['valores_negativos'] = len(valores_negativos)

    print(f"❌ Valores negativos/cero: {len(valores_negativos):,} registros")

    # 3. Área cosechada > Área sembrada
    area_inconsistente = df[df['Área cosechada (ha)'] > df['Área sembrada (ha)']]
    inconsistencias['area_logica'] = len(area_inconsistente)

    print(f"🌾 Área cosechada > sembrada: {len(area_inconsistente):,} registros")


    return inconsistencias
```

```python
def identificar_inconsistencias_formatos(df):
    """Identifica inconsistencias en los datos del dataframe agrícola"""

    print("🔍 IDENTIFICACIÓN DE INCONSISTENCIAS")
    print("=" * 50)

    inconsistencias = {}

    print(f"\n📋 ANÁLISIS DE FORMATOS:")
    for col in ['Departamento', 'Municipio', 'Cultivo']:
        if col in df.columns:
            duplicados_formato = df[col].str.strip().value_counts()
            inconsistencias = duplicados_formato[duplicados_formato.index.str.contains(' ', na=False)]
            print(f"   {col}: {len(inconsistencias)} posibles variaciones de formato")

    return inconsistencias
```

```python
def evaluar_significancia_inconsistencias(df, inconsistencias):

    print(f"\n⚖️ EVALUACIÓN DE SIGNIFICANCIA")
    print("=" * 50)

    total_registros = len(df)
    total_inconsistencias = sum(inconsistencias)
    porcentaje_total = (total_inconsistencias / total_registros) * 100

    print(f"Total de registros: {total_registros:,}")
    print(f"Total de inconsistencias: {total_inconsistencias:,}")
    print(f"Porcentaje de inconsistencias: {porcentaje_total:.2f}%")

    # Criterio de significancia
    if porcentaje_total > 10:
        recomendacion = "🚨 SÍ justifica estandarización (>5% de inconsistencias)"
    elif porcentaje_total > 3:
        recomendacion = "⚠️ Considerar estandarización (1-5% de inconsistencias)"
    else:
        recomendacion = "✅ No requiere estandarización urgente (<1% de inconsistencias)"

    print(f"\nRecomendación: {recomendacion}")
    return porcentaje_total > 1
```

####¿Qué medidas se tomaron para corregir o eliminar los datos corruptos ?

Se aplicó el siguiente proceso de limpieza para restaurar el tipo de dato correcto en las columnas afectadas:

- Conversión a Numérico Forzada: Se utilizó la función pd.to_numeric con el parámetro errors='coerce'. Esto nos permite convertir todos los valores que son números y transforma cualquier texto no numérico (como "No disponible") en un valor NaN (Not a Number), que es el marcador de nulo estándar de Pandas.

- Imputación de Valores Nulos: Todos los valores NaN que se generaron en el paso anterior fueron rellenados con el valor 0. Esta decisión se justifica bajo el supuesto de que un dato no disponible en un censo de inventario es funcionalmente equivalente a un invento de cero para ese registro, permitiendo así los cálculos agregados.

```python
print("--- Columnas REALES en inv_bufalino_df ---")
print(bufalina_df.columns)
```

```python
#Grid 3x3 correlación variables de cultivos con mayor produccion (Top 9)

def grid_correlacion(dataframe, lista_cultivos, x_variable, y_variable, titulo_principal):

    fig, axes = plt.subplots(3, 3, figsize=(18, 18))
    axes = axes.flatten()

    # Iteramos sobre cada cultivo y su eje correspondiente
    for i, cultivo in enumerate(lista_cultivos):
        ax = axes[i]

        # Filtramos los datos para el cultivo actual
        datos_cultivo = dataframe[dataframe['Cultivo'] == cultivo]

        # Creamos el gráfico de dispersión en el eje correspondiente
        sns.scatterplot(x=x_variable, y=y_variable, data=datos_cultivo, ax=ax, alpha=0.6)

        # Configuramos los títulos y etiquetas
        ax.set_title(cultivo, fontsize=14)
        ax.set_xlabel(x_variable, fontsize=10)
        ax.set_ylabel(y_variable, fontsize=10)
        ax.grid(True, linestyle='--', alpha=0.6)

    # Añadimos un título principal a toda la figura
    fig.suptitle(titulo_principal, fontsize=22, y=0.97)

    # Ajustamos el layout para evitar que los títulos se superpongan
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
# --- Función para crear un grid de gráficos de correlación 3x3 ---

def correlacion_scatterplot(dataframe, lista_cultivos, x_variable, y_variable, titulo_principal):

    fig, axes = plt.subplots(3, 3, figsize=(18, 18))

    axes = axes.flatten()

    # Iteramos sobre cada cultivo y su eje correspondiente
    for i, cultivo in enumerate(lista_cultivos):
        ax = axes[i]

        datos_cultivo = dataframe[dataframe['Cultivo'] == cultivo]

        # Creamos el gráfico de dispersión en el eje correspondiente
        sns.scatterplot(x=x_variable, y=y_variable, data=datos_cultivo, ax=ax, alpha=0.6)

        # Configuramos los títulos y etiquetas
        ax.set_title(cultivo, fontsize=14)
        ax.set_xlabel(x_variable, fontsize=10)
        ax.set_ylabel(y_variable, fontsize=10)
        ax.grid(True, linestyle='--', alpha=0.6)

    fig.suptitle(titulo_principal, fontsize=22, y=0.97)

    plt.tight_layout(rect=[0, 0.03, 1, 0.95])
    plt.show()
```

```python
# --- GRID 3x3 DE ANÁLISIS DE REGRESIÓN ---

fig, axes = plt.subplots(3, 3, figsize=(22, 22))

axes = axes.flatten()

# Iteramos sobre cada cultivo y su eje correspondiente
for i, cultivo in enumerate(top_9_cultivos):
    ax = axes[i]

    # Filtramos y realizamos la regresión
    datos_cultivo = agricola_df[agricola_df['Cultivo'] == cultivo]
    x_var = datos_cultivo['Área cosechada (ha)']
    y_var = datos_cultivo['Producción (t)']

    if len(x_var) > 1:
        pend, interc, corr, p_valor, _ = stats.linregress(x_var, y_var)
        r_cuadrado = corr**2
        y_pred = pend * x_var + interc

        ax.scatter(x_var, y_var, alpha=0.5, label='Datos Reales')
        ax.plot(x_var, y_pred, color='red', linewidth=2, label='Línea de Regresión')
        ax.set_title(f'{cultivo}\n(R² = {r_cuadrado:.3f})', fontsize=14)
    else:
        ax.set_title(f'{cultivo}\n(Datos insuficientes)', fontsize=14)

    # Añadimos etiquetas a CADA gráfico
    ax.set_xlabel('Área Cosechada (ha)', fontsize=10)
    ax.set_ylabel('Producción (t)', fontsize=10)
    ax.grid(True, linestyle='--')

fig.suptitle('Análisis de Regresión por Cultivo Principal', fontsize=24)
plt.tight_layout(rect=[0, 0.03, 1, 0.96])
plt.show()
```

Sí, es totalmente posible y, como demuestra el análisis por cultivo, La fuerte correlación positiva entre **`Área cosechada (ha)`** y **`Producción (t)`**, observada de manera general, se confirma para cada uno de los 9 cultivos principales, aunque con matices importantes.

El grid de regresión 3x3 nos permite cuantificar la fuerza de esta relación para cada cultivo a través del **coeficiente de determinación (R²)**:

*   **Modelos de Alta Predictibilidad (R² > 0.90):** Para cultivos como la **Papa (R² = 0.944)**, **Arroz (R² = 0.962)**, **Banano (R² = 0.948)**, **Palma de aceite (R² = 0.923)** y **Caña (R² = 0.910)**, el área cosechada por sí sola puede explicar más del 90% de la variabilidad en la producción. Esto indica que para estos cultivos, la producción es altamente predecible a partir del área.

*   **Modelos de Buena Predictibilidad (0.80 < R² < 0.90):** Para cultivos como el **Plátano (R² = 0.805)**, **Yuca (R² = 0.857)**, **Maíz (R² = 0.859)** y **Aguacate (R² = 0.824)**, la relación sigue siendo muy fuerte y estadísticamente significativa. Sin embargo, el R² ligeramente menor sugiere que, aunque el área cosechada es el factor principal, otras variables (como el rendimiento, las condiciones climáticas, la tecnología, etc.) tienen un impacto mayor en la producción final en comparación con el primer grupo.

#### **¿Cuáles fueron las variables seleccionadas para el análisis de regresiones y cómo se eligieron en función de los objetivos del proyecto?**



Para el análisis de regresiones se realizaron dos modelos complementarios, cada uno diseñado para investigar una etapa específica del ciclo agrícola. La selección de variables se basó en una lógica de causa y efecto para descomponer y entender el proceso productivo en su totalidad.

**1. Modelo de Producción (Resultado Final):**
Este primer análisis modela la relación entre el esfuerzo efectivo y el resultado final.

*   **Variable Dependiente (A estimar): `Producción (t)`**. Se seleccionó como la variable de interés a predecir, ya que la producción en toneladas es el objetivo principal y la métrica de éxito del proceso agrícola.

**2. Modelo de Eficiencia del Proceso (Riesgo y Pérdida):**
Este segundo análisis modela la eficiencia y predictibilidad de la fase inicial del cultivo.
*   **Variable Independiente (Predictora): `Área sembrada (ha)`**. Se eligió esta variable porque representa la inversión inicial, el esfuerzo planificado y el punto de partida del ciclo.
*   **Variable Dependiente (A estimar): `Área cosechada (ha)`**. En este modelo, el área cosechada se convierte en la variable de resultado. Se seleccionó para cuantificar qué porción del esfuerzo inicial sobrevive y se convierte en un esfuerzo efectivo, midiendo así la pérdida o el riesgo inherente al cultivo.

#### **5.1.3. Estadística inferencial (Opcional)**
---

- ¿Resulta pertinente calcular intervalos de confianza para estimar algún parámetro con confianza a partir de una muestra?
- ¿Es posible realizar pruebas de hipótesis para probar alguna teoría sobre un parámetro de la población?

Con el fin de complementar el análisis descriptivo y correlacional, se aplicaron técnicas de estadística inferencial sobre la variable Rendimiento (t/ha):

Intervalo de confianza (IC 95%)
El rendimiento promedio estimado para el conjunto de datos es de 10.60 t/ha.
El intervalo de confianza al 95% se encuentra entre [10.52, 10.68] t/ha, lo cual indica que, con un 95% de confianza, el verdadero rendimiento promedio poblacional se ubica en dicho rango. El tamaño de la muestra (141,073 observaciones) otorga una alta precisión a esta estimación.

Prueba de hipótesis
Se evaluó la hipótesis nula:

H0: μ = 10 t/ha (el rendimiento promedio poblacional es igual a 10 t/ha).

H1: μ ≠ 10 t/ha (el rendimiento promedio poblacional es diferente de 10 t/ha).

El resultado arrojó un estadístico t = 14.44 y un p-valor < 0.001, lo que permite rechazar la hipótesis nula. Por lo tanto, existe evidencia estadísticamente significativa de que el rendimiento promedio es superior a 10 t/ha.

```python
### Intervalos de confianza (IC) para el rendimiento promedio ###
# Tomamos la variable rendimiento (t/ha)
rendimiento = agricola_df['Rendimiento (t/ha)'].dropna()


media_rend = np.mean(rendimiento)
std_rend = np.std(rendimiento, ddof=1)  # ddof=1 para usar varianza muestral
n = len(rendimiento)

# Nivel de confianza 95%
alpha = 0.05
t_critico = stats.t.ppf(1 - alpha/2, df=n-1)  # valor t
margen_error = t_critico * (std_rend / np.sqrt(n))

ic_inf = media_rend - margen_error
ic_sup = media_rend + margen_error

print("Intervalo de Confianza del 95% para el Rendimiento Promedio (t/ha):")
print(f"[{ic_inf:.2f}, {ic_sup:.2f}]")
print(f"Media: {media_rend:.2f}, Tamaño de muestra: {n}")

### 2. Prueba de hipótesis ###
# Ejemplo: H0 = el rendimiento promedio es 10 t/ha
hipotesis = 10
t_stat, p_valor = stats.ttest_1samp(rendimiento, hipotesis)

print("\nPrueba de Hipótesis sobre el Rendimiento Promedio (t/ha):")
print(f"H0: μ = {hipotesis}")
print(f"t = {t_stat:.3f}, p-valor = {p_valor:.4f}")

if p_valor < 0.05:
    print("Se rechaza H0: el rendimiento promedio es significativamente diferente de 10 t/ha.")
else:
    print("No se rechaza H0: no hay evidencia suficiente para decir que el rendimiento promedio difiere de 10 t/ha.")
```

```python
# Simular datos mensuales basados en los datos anuales
def crear_datos_mensuales(df):
    data_mensual = []
    for _, row in df.iterrows():
        año = row['Año']
        for mes in range(1, 13):
            # Simular progreso acumulativo mensual
            factor_mes = mes / 12.0
            # Agregar variación estacional (más producción en meses de cosecha)
            if mes in [3, 4, 5, 10, 11]:  # Meses de cosecha típicos
                factor_mes *= 1.2
            elif mes in [1, 2, 12]:  # Meses de menor actividad
                factor_mes *= 0.8

            data_mensual.append({
                'Departamento': row['Departamento'],
                'Cultivo': row['Cultivo'],
                'Año': año,
                'Mes': mes,
                'Fecha': f"{año}-{mes:02d}",
                'Producción_Acumulada': row['Producción (t)'] * factor_mes,
                'Área_Acumulada': row['Área cosechada (ha)'] * factor_mes,
                'Rendimiento': row['Rendimiento (t/ha)']
            })
    return pd.DataFrame(data_mensual)
```

```python

"""Gráfico de progreso mensual por cultivo y departamento"""

# Crear datos mensuales
data_mensual = crear_datos_mensuales(agricola_df)

# Filtrar top cultivos y departamentos para mejor visualización
top_cultivos = agricola_df.groupby('Cultivo')['Producción (t)'].sum().nlargest(6).index
top_deptos = agricola_df.groupby('Departamento')['Producción (t)'].sum().nlargest(8).index

data_filtrada = data_mensual[
    (data_mensual['Cultivo'].isin(top_cultivos)) &
    (data_mensual['Departamento'].isin(top_deptos))
]

# Crear gráfico animado por mes
fig = px.scatter(data_filtrada,
                x='Área_Acumulada',
                y='Producción_Acumulada',
                size='Rendimiento',
                color='Departamento',
                hover_name='Cultivo',
                animation_frame='Fecha',
                title='📊 Progreso Mensual de Producción por Cultivo y Departamento',
                labels={
                    'Área_Acumulada': 'Área Cosechada Acumulada (ha)',
                    'Producción_Acumulada': 'Producción Acumulada (t)',
                    'Rendimiento': 'Rendimiento (t/ha)'
                })

fig.update_layout(
    height=600,
    showlegend=True,
    title_font_size=16,
    xaxis_title_font_size=14,
    yaxis_title_font_size=14
)

# Configurar animación más lenta para mejor visualización
fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 800
fig.layout.updatemenus[0].buttons[0].args[1]['transition']['duration'] = 300

fig.show()
```

```python
"""Mapa de calor por cultivo y departamento"""

# Preparar matriz de datos
top_dptos = agricola_df.groupby('Departamento')['Producción (t)'].sum().nlargest(12).index
top_cultivos = agricola_df.groupby('Cultivo')['Producción (t)'].sum().nlargest(15).index

# Crear matriz de rendimiento
matriz_data = []
for dpto in top_dptos:
    for cultivo in top_cultivos:
        subset = agricola_df[
            (agricola_df['Departamento'] == dpto) &
            (agricola_df['Cultivo'] == cultivo)
        ]
        if len(subset) > 0:
            rend_promedio = subset['Rendimiento (t/ha)'].mean()
            prod_total = subset['Producción (t)'].sum()
            area_total = subset['Área cosechada (ha)'].sum()
            registros = len(subset)
        else:
            rend_promedio = 0
            prod_total = 0
            area_total = 0
            registros = 0

        matriz_data.append({
            'Departamento': dpto,
            'Cultivo': cultivo,
            'Rendimiento': rend_promedio,
            'Producción': prod_total,
            'Área': area_total,
            'Registros': registros
        })

matriz_df = pd.DataFrame(matriz_data)
matriz_pivot = matriz_df.pivot(index='Departamento', columns='Cultivo', values='Rendimiento')

# Crear heatmap interactivo
fig = go.Figure(data=go.Heatmap(
    z=matriz_pivot.values,
    x=matriz_pivot.columns,
    y=matriz_pivot.index,
    colorscale='RdYlBu_r',
    hoverongaps=False,
    hovertemplate='<b>%{y}</b><br>' +
                  '<b>%{x}</b><br>' +
                  'Rendimiento: %{z:.2f} t/ha<br>' +
                  '<extra></extra>'
))

fig.update_layout(
    title='🔥 Mapa de Calor: Rendimiento por Departamento y Cultivo',
    xaxis_title='Cultivos',
    yaxis_title='Departamentos',
    height=700,
    xaxis={'tickangle': 45},
    font=dict(size=10)
)

fig.show()
```

## **6. Analitica Inferencial**

- ¿El conjunto de datos tiene una variable objetivo a estimar?, de ser así, ¿es una variable continúa o categórica?

```python
# 1. Identificación de variables objetivo
print('=== VARIABLES OBJETIVO POTENCIALES ===')
print('Según los objetivos del proyecto:')
print('1. Crecimiento/Decrecimiento (categórica): Clasificar cultivos según su tendencia')
print('2. Rendimiento (continua): Predecir productividad por hectárea')
print('3. Producción (continua): Estimar producción total')

# Crear variable categórica de crecimiento
# Calcular cambio porcentual entre 2023 y 2024 por cultivo y municipio
agricola_2023 = agricola_geo[agricola_geo['Año'] == 2023].groupby(['Código Dane municipio', 'Cultivo']).agg({
    'Producción (t)': 'sum',
    'Área sembrada (ha)': 'sum'
}).reset_index()

agricola_2024 = agricola_geo[agricola_geo['Año'] == 2024].groupby(['Código Dane municipio', 'Cultivo']).agg({
    'Producción (t)': 'sum',
    'Área sembrada (ha)': 'sum'
}).reset_index()

# Merge para comparar
comparacion = pd.merge(agricola_2023, agricola_2024, on=['Código Dane municipio', 'Cultivo'], suffixes=('_2023', '_2024'))

# Calcular cambio porcentual
comparacion['cambio_produccion'] = ((comparacion['Producción (t)_2024'] - comparacion['Producción (t)_2023']) / comparacion['Producción (t)_2023']) * 100

# Crear variable categórica
def clasificar_crecimiento(cambio):
    if cambio > 10: return 'Alto_Crecimiento'
    elif cambio > 0: return 'Crecimiento_Moderado'
    elif cambio > -10: return 'Estable'
    else: return 'Decrecimiento'

comparacion['categoria_crecimiento'] = comparacion['cambio_produccion'].apply(clasificar_crecimiento)

print('\nDistribución de categorías de crecimiento:')
print(comparacion['categoria_crecimiento'].value_counts())

print('Variables objetivo identificadas:')
print('- categoria_crecimiento: Variable categórica (4 clases)')
print('- Rendimiento (t/ha): Variable continua')
print('- Producción (t): Variable continua')
```

**Respuesta:** Sí, el dataset tiene múltiples variables objetivo potenciales:
- **Categórica**: Categoría de crecimiento (Alto_Crecimiento, Crecimiento_Moderado, Estable, Decrecimiento)
- **Continuas**: Rendimiento (t/ha) y Producción (t) para modelos de regresión

Los objetivos del proyecto incluyen:
1. Identificar cambios significativos entre 2023 y 2024 en área sembrada, producción y rendimiento
2. Incorporar análisis del inventario pecuario bovino para visión integral del sector agropecuario

```python
# 2. Análisis de desbalanceo de datos
print('=== ANÁLISIS DE DESBALANCEO ===')

# Distribución de categorías de crecimiento
print('Distribución de categorías de crecimiento:')
dist_crecimiento = comparacion['categoria_crecimiento'].value_counts()
print(dist_crecimiento)
print('\nPorcentajes:')
print((dist_crecimiento / len(comparacion) * 100).round(2))

# Distribución por grupos de cultivos
print('\n=== DISTRIBUCIÓN POR GRUPOS DE CULTIVOS ===')
dist_grupos = agricola_geo['Grupo cultivo'].value_counts()
print(dist_grupos)
print('\nPorcentajes:')
print((dist_grupos / len(agricola_geo) * 100).round(2))

# Distribución por años
print('\n=== DISTRIBUCIÓN TEMPORAL ===')
dist_anos = agricola_geo['Año'].value_counts().sort_index()
print(dist_anos)
print('\nPorcentajes:')
print((dist_anos / len(agricola_geo) * 100).round(2))

# Distribución por departamentos (top 10)
print('\n=== TOP 10 DEPARTAMENTOS ===')
dist_deptos = agricola_geo['Departamento'].value_counts().head(10)
print(dist_deptos)

# Cálculo de índice de desbalanceo
def calcular_desbalanceo(serie):
    counts = serie.value_counts()
    return counts.max() / counts.min()

print('\n=== ÍNDICES DE DESBALANCEO ===')
print(f'Categorías de crecimiento: {calcular_desbalanceo(comparacion["categoria_crecimiento"]):.2f}')
print(f'Grupos de cultivos: {calcular_desbalanceo(agricola_geo["Grupo cultivo"]):.2f}')
print(f'Años: {calcular_desbalanceo(agricola_geo["Año"]):.2f}')
print('\nNota: Índice > 3 indica desbalanceo significativo')
```

**Respuesta:** Se identifica desbalanceo moderado en varias dimensiones:

- **Grupos de cultivos**: Desbalanceo significativo, con predominio de ciertos grupos como cereales y frutales
- **Distribución temporal**: Balanceada entre 2023 y 2024
- **Distribución geográfica**: Concentración en departamentos con mayor actividad agrícola

El desbalanceo es esperado y refleja la realidad del sector agropecuario colombiano.
